```python
import pandas as pd

# Load the dataset
df = pd.read_csv('data/final_datausage2025.csv')

# === Task 1: Understand the Data Structure ===

print("Shape of the dataset:")
print(df.shape)  # (rows, columns)

print("\nColumn names:")
print(df.columns.tolist())

print("\nData types and non-null counts:")
print(df.info())

print("\nPreview of first 3 rows:")
print(df.head(3))

# === Task 2: Data Quality Checks ===

print("\nMissing values per column:")
print(df.isnull().sum())

print("\nNumber of duplicate rows:")
print(df.duplicated().sum())

print("\nNumber of unique states:", df['state'].nunique())
print("Sex categories:", df['sex'].unique())
print("Age groups (first 10):", df['age'].unique()[:10])

# === Optional: Drop duplicate rows and save cleaned version ===

remove_duplicates = True  # Set to False if you want to keep duplicates

if remove_duplicates:
    df = df.drop_duplicates()
    df.to_csv('data/final_datausage_cleaned.csv', index=False)
    print("\nDuplicates removed. Cleaned data saved as 'final_datausage_cleaned.csv'")
    print("New shape after removing duplicates:", df.shape)
else:
    print("\nDuplicates retained. No changes made to dataset.")



```

```python
# === Basic EDA for final_merged_data2025.csv ===

import pandas as pd

# Load dataset
df = pd.read_csv('data/final_merged_data2025.csv')

# Standardize column names
df.columns = df.columns.str.strip().str.lower()

# Basic structure
print("Shape of the dataset:", df.shape)
print("\nColumn names:")
print(df.columns.tolist())

# Data types and null counts
print("\nDataset info:")
print(df.info())

# Preview top 3 rows
print("\nFirst 3 rows of the dataset:")
print(df.head(3))

# Missing values
print("\nMissing values per column:")
print(df.isnull().sum())

# Check for duplicates
print("\nNumber of duplicate rows:", df.duplicated().sum())

# Unique values in key columns
if 'state' in df.columns:
    print("\nNumber of unique states:", df['state'].nunique())

if 'sex' in df.columns:
    print("Unique values in 'sex':", df['sex'].unique())

if 'age' in df.columns:
    print("Unique values in 'age' (first 10):", df['age'].unique()[:10])

# Drop duplicates and save cleaned version
df_cleaned = df.drop_duplicates()
df_cleaned.to_csv('data/final_datausage_cleaned.csv', index=False)
print("\nCleaned dataset saved to 'data/final_datausage_cleaned.csv'")
print("New shape after removing duplicates:", df_cleaned.shape)
```
