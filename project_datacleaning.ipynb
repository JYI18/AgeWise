```python
import pandas as pd

# Load the dataset
df = pd.read_csv("data/urban-development.csv")

# Remove the first row (redundant header)
df = df.iloc[1:].copy()

# Strip whitespace
df.columns = df.columns.str.strip()
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)

# Convert data types
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')
df['Value'] = pd.to_numeric(df['Value'], errors='coerce')

# Check for missing values
print(df.isnull().sum())

# Preview cleaned data
print(df.head())

# Optional: Save to new CSV
df.to_csv("urban-clean.csv", index=False)
```

```python
import pandas as pd

# Load the urban2 dataset
df = pd.read_csv("data/urban2.csv")

# Clean column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# Trim string columns
df['strata'] = df['strata'].str.strip().astype('category')
df['main_groups'] = df['main_groups'].str.strip().astype('category')

# Convert year to integer (in case it's read as object)
df['year'] = df['year'].astype(int)

# Convert index number to float
df['index_number'] = df['index_number'].astype(float)

# Optional: sort by year, strata, and main_groups
df = df.sort_values(by=['year', 'strata', 'main_groups'])

# Save cleaned version
df.to_csv("data/urban2_cleaned.csv", index=False)
print("Saved")

# Preview result
print(df.head())
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('data/urban-development.csv')

# Clean column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# Strip and convert string columns to category (if present)
if 'strata' in df.columns:
    df['strata'] = df['strata'].astype(str).str.strip().astype('category')

if 'main_groups' in df.columns:
    df['main_groups'] = df['main_groups'].astype(str).str.strip().astype('category')

# Convert year safely to integer
if 'year' in df.columns:
    df['year'] = pd.to_numeric(df['year'], errors='coerce').astype('Int64')

# Convert index_number safely to float
if 'index_number' in df.columns:
    df['index_number'] = pd.to_numeric(df['index_number'], errors='coerce')

# Optional: drop rows with missing year or index values (if needed)
df = df.dropna(subset=['year'])  # or ['year', 'index_number'] if both required

# Sort (if columns are present)
sort_cols = [col for col in ['year', 'strata', 'main_groups'] if col in df.columns]
df = df.sort_values(by=sort_cols)

# Save cleaned version
df.to_csv('data/cleaned_urban_development.csv', index=False)

# Confirm
print(" Saved ")
print(df.head())
```

```python
import pandas as pd

# Load the CSV with no header
df = pd.read_csv('data/rural_urban.csv', header=None, names=['year', 'urban_population'])

# Clean column names to snake_case (already done in names=)
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# Convert data types
df['year'] = pd.to_numeric(df['year'], errors='coerce').astype('Int64')
df['urban_population'] = pd.to_numeric(df['urban_population'], errors='coerce').astype('Int64')

# Drop rows with missing year or population (if any)
df = df.dropna(subset=['year', 'urban_population'])

# Save cleaned file
df.to_csv('data/rural_urban_cleaned.csv', index=False)

# Preview
print(" Saved ")
print(df.head())
```

```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('data/urban_rural.csv')  # Replace with your actual filename

# Clean column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# Strip and clean text columns
df['state'] = df['state'].astype(str).str.strip().astype('category')
df['strata'] = df['strata'].astype(str).str.strip().astype('category')

# Convert types
df['year'] = pd.to_numeric(df['year'], errors='coerce').astype('Int64')
df['percentage'] = pd.to_numeric(df['percentage'], errors='coerce')

# Drop rows with missing values (if any)
df = df.dropna(subset=['year', 'state', 'strata', 'percentage'])

# Sort for neatness
df = df.sort_values(by=['year', 'state', 'strata'])

# Save cleaned version
df.to_csv('data/urban_rural_cleaned.csv', index=False)

# Preview
print("Saved")
print(df.head())
```

```python
import pandas as pd

# Load preprocessed datasets
df_index = pd.read_csv('data/cleaned_urban_development.csv')
df_development = pd.read_csv('data/urban_rural_cleaned.csv')
df_population = pd.read_csv('data/rural_urban_cleaned.csv')
df_state_pct = pd.read_csv('data/urban2_cleaned.csv')

# ----------------------------
# Step 1: Inner Merge (Test Alignment)
# ----------------------------
merged_test = pd.merge(df_index, df_development, on='year', how='inner')
merged_test = pd.merge(merged_test, df_population, on='year', how='inner')
merged_test = pd.merge(merged_test, df_state_pct, on='year', how='inner')

print("Test Merge Complete")
print("Years in common across all datasets:", merged_test['year'].unique())
print(merged_test.info())

# ----------------------------
# Step 2: Full Outer Merge
# ----------------------------
merged_all = pd.merge(df_index, df_development, on='year', how='outer')
merged_all = pd.merge(merged_all, df_population, on='year', how='outer')
merged_all = pd.merge(merged_all, df_state_pct, on='year', how='outer')

# Optional: Sort by year
merged_all = merged_all.sort_values(by='year')

# Save final merged dataset
merged_all.to_csv('data/merged_urban_data.csv', index=False)

# Preview
print(" Final merged dataset saved as merged_urban_data.csv")
print(merged_all.head())
```
